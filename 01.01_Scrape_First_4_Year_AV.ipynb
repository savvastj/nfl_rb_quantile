```python
import pandas as pd
import pfr
```

```python
url_template = "https://www.pro-football-reference.com/play-index/psl_finder."\
               "cgi?request=1&match=combined&year_min=2000&year_max=2017&seas"\
               "on_start=1&season_end=4&pos%5B%5D=qb&pos%5B%5D=rb&pos%5B%5D=wr"\
               "&pos%5B%5D=te&pos%5B%5D=e&pos%5B%5D=t&pos%5B%5D=g&pos%5B%5D=c&"\
               "pos%5B%5D=ol&pos%5B%5D=dt&pos%5B%5D=de&pos%5B%5D=dl&pos%5B%5D="\
               "ilb&pos%5B%5D=olb&pos%5B%5D=lb&pos%5B%5D=cb&pos%5B%5D=s&pos%5B"\
               "%5D=db&pos%5B%5D=k&pos%5B%5D=p&draft_year_min=1936&draft_year_"\
               "max=2017&draft_slot_min=1&draft_slot_max=500&draft_pick_in_"\
               "round=pick_overall&conference=any&draft_pos%5B%5D=qb&draft_pos"\
               "%5B%5D=rb&draft_pos%5B%5D=wr&draft_pos%5B%5D=te&draft_pos%5B%5"\
               "D=e&draft_pos%5B%5D=t&draft_pos%5B%5D=g&draft_pos%5B%5D=c&"\
               "draft_pos%5B%5D=ol&draft_pos%5B%5D=dt&draft_pos%5B%5D=de&draft"\
               "_pos%5B%5D=dl&draft_pos%5B%5D=ilb&draft_pos%5B%5D=olb&"\
               "draft_pos%5B%5D=lb&draft_pos%5B%5D=cb&draft_pos%5B%5D=s&"\
               "draft_pos%5B%5D=db&draft_pos%5B%5D=k&draft_pos%5B%5D=p&c5val="\
               "1.0&order_by=av&offset={}"
```

```python
row_css_selector = "#results > tbody > tr" 
col_css_selector = "#results > thead > tr:nth-child(2) > th"
dfs = []
errors = []
# i's represent page offset values
for i in range (0, 8001, 100):
    url = url_template.format(i)
    try:
        df = pfr.get_table(url, row_css_selector, col_css_selector, 1)
        dfs.append(df)
        print("Scraped:", i)
    except Exception as e:
        errors.append([url, e])
        print("ERROR:", i)
```

```python
df = pd.concat(dfs, ignore_index=True)
```
