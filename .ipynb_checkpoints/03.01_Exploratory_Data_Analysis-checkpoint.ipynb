### Nullity correlation matrix, note from the package doc:
> Nullity correlation ranges from -1 (if one variable appears the other definitely does not) to 0 (variables appearing or not appearing have no effect on one another) to 1 (if one variable appears the other definitely also does).

> Variables that are always full or always empty have no meaningful correlation, and so are silently removed from the visualization

```python
# Nullity correlation matrix for variables with missing data
msno.heatmap(train_df);
```

### Dendrogram

> To interpret this graph, read it from a top-down perspective. Cluster leaves which linked together at a distance of zero fully predict one another's presenceone variable might always be empty when another is filled, or they might always both be filled or both empty, and so on.

```python
msno.dendrogram(train_df);
```

## Print out exact Pearson correlation values

```python
av_pearson_corr = train_df[features].corr().AV.sort_values()
av_pearson_corr
```

```python
av_spearman_corr = train_df[features].corr('spearman').AV.sort_values()
av_spearman_corr
```

```python
sns.set(style='white', palette='colorblind', font_scale=1.3, 
        rc={'figure.figsize': (12, 9)})

# Set up the the 5x3 plotting grid by creating figure and axes objects
# Set sharey to True so that each row of plots share the left most y-axis labels
nrows = 9
ncols = 4
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, sharey=True,
                         figsize=(12, 24))

for feat, ax in zip(features[:-1], axes.flatten()):
    sns.regplot(x=feat, y='AV', data=train_df, lowess=True, ax=ax, 
                scatter_kws={'alpha': 0.5, 'color': 'salmon'},
                line_kws={'color': 'k'})
    ax.set_xlabel(feat)
    

# get rid of blank plots
    for i in range(len(features[:-1]), nrows*ncols):
        axes.flatten()[i].axis('off')

# tighten up the padding for the subplots
fig.tight_layout()
```

```python
sns.regplot(x=train_df.Wt / train_df.Forty, y=train_df.AV, lowess=True)
plt.xlabel('Weight to Forty Ratio')
```

```python
sns.regplot(x=train_df.Wt * train_df.Forty, y=train_df.AV, lowess=True)
plt.xlabel("Weight and Forty Interaction");
```

```python
nrows = 8
ncols = 4
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, sharey=True,
                         figsize=(12, 20))
feats_to_plot = [feat for feat in features if feat not in ['AV', 'Wt']]
for feat, ax in zip(feats_to_plot, axes.flatten()):
    wt_to_feat_ratio =  train_df.Wt / train_df[feat]
    sns.regplot(x=wt_to_feat_ratio, y=train_df.AV, lowess=True, ax=ax, 
                scatter_kws={'alpha': 0.5, 'color': 'salmon'},
                line_kws={'color': 'k'})
    ax.set_xlabel('Wt / ' + feat )
    

# get rid of blank plots
    for i in range(len(feats_to_plot), nrows*ncols):
        axes.flatten()[i].axis('off')

# tighten up the padding for the subplots
fig.tight_layout()
```

# Correlations for all feature ratios and feature interactions
## Ratios

```python
import itertools
```

```python
ratio_df = train_df.copy()
ratio_pairs = itertools.combinations(ratio_df[features[:-1]].columns, 2)
for pair in ratio_pairs:
    new_col = ' / '.join(pair)
    ratio_df[new_col] = ratio_df[pair[0]] / ratio_df[pair[1]]

ratio_pair_names = ratio_df.columns[ratio_df.columns.str.contains('/')].tolist()

ratio_pair_names += ['AV']
```

```python
spearman_ratio_corr = ratio_df[ratio_pair_names].corr('spearman')

av_spearman_ratio_corr = (spearman_ratio_corr.reindex(spearman_ratio_corr.AV
                                                       .abs()
                                                       .sort_values(ascending=False)
                                                       .index)
                                             .AV)
```

```python
av_spearman_ratio_corr.head(30)                        
```

```python
pearson_ratio_corr = ratio_df[ratio_pair_names].corr('pearson')

av_pearson_ratio_corr = (pearson_ratio_corr.reindex(pearson_ratio_corr.AV
                                                       .abs()
                                                       .sort_values(ascending=False)
                                                       .index)
                                             .AV)
```

```python
av_pearson_ratio_corr.head(30)
```

## Interactions

```python
inter_df = train_df.copy()
inter_pairs = itertools.combinations(inter_df[features[:-1]].columns, 2)
for pair in inter_pairs:
    new_col = ' * '.join(pair)
    inter_df[new_col] = inter_df[pair[0]] * inter_df[pair[1]]

inter_pair_names = inter_df.columns[inter_df.columns.str.contains('\*')].tolist()

inter_pair_names += ['AV']
```

```python
spearman_inter_corr = inter_df[inter_pair_names].corr('spearman')

av_spearman_inter_corr = (spearman_inter_corr.reindex(spearman_inter_corr.AV
                                                       .abs()
                                                       .sort_values(ascending=False)
                                                       .index)
                                             .AV)
```

```python
av_spearman_inter_corr.head(30)
```

```python
pearson_inter_corr = inter_df[inter_pair_names].corr('pearson')

av_pearson_inter_corr = (pearson_inter_corr.reindex(pearson_inter_corr.AV
                                                       .abs()
                                                       .sort_values(ascending=False)
                                                       .index)
                                             .AV)
```
