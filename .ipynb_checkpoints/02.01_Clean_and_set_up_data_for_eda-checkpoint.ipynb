# Clean up combine data

```python
combine_df = pd.read_csv("raw_data/pfr_combine_data.csv")
```

```python
combine_df.info()
```

```python
combine_df.head()
```

```python
# a bunch cleaning, get team that drafted player, format height to inches
# drop some columns, and rename others so they play nicer with pandas
combine_df["Team_Drafted"] = (combine_df["Drafted (tm/rnd/yr)"]
                                  .str.split(" / ", expand=True)[0])
ht = combine_df.Ht.str.split("-", expand=True).astype(int)
combine_df["Ht"] = ht[0]*12 + ht[1]
combine_df.drop(["College", "Drafted (tm/rnd/yr)"], axis=1, inplace=True)
combine_df.rename(columns={"40yd": "Forty", "3Cone": "Cone", 
                           "School": "College"}, inplace=True)
```

```python
combine_df.info()
```

```python
combine_df.head()
```

```python
combine_df.to_csv("processed_data/pfr_combine_data.csv", index=False)
```

# Set up RB college stats

```python
rush_rec_df = pd.read_csv("processed_data/rb_yearly_college_rush_rec_stats.csv")
```

```python
rush_rec_df.info()
```

```python
rush_rec_df.loc[rush_rec_df.Pos.isnull()]
```

```python
rush_rec_df.loc[rush_rec_df.Pos.isnull(), 'Pos'] = 'RB'
```

```python
# fill nulls in numeric columns with 0s
rush_rec_df.loc[:, 'G':'Scrim_TD'] = rush_rec_df.loc[:, 'G':'Scrim_TD'].fillna(0)
```

```python
# grpy id, then get to totals for each stat plus the number of years played
rush_rec_grpby_player = rush_rec_df.groupby("Sref_Cfb_ID")
rush_rec_cols = ["G", "Rush_Att", "Rush_Yds", "Rush_TD", "Rec", "Rec_Yds", "Rec_TD",
                 "Plays", "Scrim_Yds", "Scrim_TD"]
f = {col: "sum" for col in rush_rec_cols}
f["Year"] = ["count", "max"]
rush_rec_totals = rush_rec_grpby_player.agg(f).reset_index()
```

```python
# edit columns
cols = rush_rec_totals.columns.droplevel(1)[:-2].tolist() +\
       ["Col_Yrs_Played", "Col_Last_Yr_Played"]
rush_rec_totals.columns = cols
```

```python
rush_rec_totals.info()
```

```python
rush_rec_totals.loc[rush_rec_totals.Rec==0, 'Rec_Yds'] / rush_rec_totals.loc[rush_rec_totals.Rec==0, 'Rec']
```

```python
# construct more features
rush_rec_totals["Rush_Yds_per_Att"] = (rush_rec_totals.Rush_Yds / rush_rec_totals.Rush_Att).fillna(0)
rush_rec_totals["Rec_Yds_per_Rec"] = (rush_rec_totals.Rec_Yds / rush_rec_totals.Rec).fillna(0)
rush_rec_totals["Scrim_Avg"] = (rush_rec_totals.Scrim_Yds / rush_rec_totals.Plays).fillna(0)

rush_rec_per_g = (rush_rec_totals.loc[:, "Rush_Att":"Scrim_TD"]
                                  .div(rush_rec_totals.G, axis=0))
rush_rec_per_g_cols = [col+"_per_G" for col in rush_rec_per_g.columns]
rush_rec_totals[rush_rec_per_g_cols] = rush_rec_per_g
```

```python
rush_rec_totals.info()
```

```python
# double check to see that there are no nulls for avg yards per rec
rush_rec_totals.loc[rush_rec_totals.Rec==0]
```

```python
pd.options.display.max_columns = None
```

```python
rush_rec_totals.head()
```

```python
rush_rec_totals.to_csv("processed_data/rb_total_college_rush_and_rec_stats.csv",
                       index=False)
```

# Finally merge the combine data and career college stats for RBs

```python
av_df = pd.read_csv("raw_data/first_4_season_AV_since_2000.csv")
```

```python
av_df.info()
```

```python
av_df.head()
```

```python
rb_df = (combine_df.query("Pos == 'RB'")
                   .merge(rush_rec_totals, on="Sref_Cfb_ID", how="left")
                   .merge(av_df[["Pfr_ID", "AV"]], on="Pfr_ID", how="left"))
```

```python
rb_df.info()
```

## Some players don't have college games but have college player ids which is strange

```python
null_gms_and_has_sref_id = (rb_df.G.isnull()) & (rb_df.Sref_Cfb_ID.notnull())
```

```python
rb_df.loc[null_gms_and_has_sref_id]
```

### Looks like the Sref_Cfb_ID scraped from the combine web pages are incorrect, so need to correct the above IDs and fill in the missing data (this is will be done 'manually')

```python
# set proper sref ids
rb_df.loc[null_gms_and_has_sref_id, "Sref_Cfb_ID"] = ["brian-calhoun-1",
                                                      "roy-helu-jr-1",
                                                      "charles-sims-1"]
```

```python
rb_df.loc[null_gms_and_has_sref_id]
```

```python
# fill in stats
stats = [[14+12+13, 619, 2760, 27, 90, 909, 4, 709, 3669, 31, 3, 2005],
         [7+12+14+14, 578, 3404, 28, 54, 501, 0, 632, 3905, 28, 4, 2010],
         [14+13+9+12, 208, 1095, 11, 45, 401, 3, 253, 1496, 14, 4, 2013]]

rb_df.loc[null_gms_and_has_sref_id, "G":"Col_Last_Yr_Played"] = stats
```

```python
rb_df.loc[null_gms_and_has_sref_id]
```

```python
rb_df.loc[null_gms_and_has_sref_id, "Rush_Yds_per_Att"] = (rb_df.loc[null_gms_and_has_sref_id, "Rush_Yds"] / 
                                                           rb_df.loc[null_gms_and_has_sref_id, "Rush_Att"])
rb_df.loc[null_gms_and_has_sref_id, "Rec_Yds_per_Rec"] = (rb_df.loc[null_gms_and_has_sref_id, "Rec_Yds"] / 
                                                      rb_df.loc[null_gms_and_has_sref_id, "Rec"])
rb_df.loc[null_gms_and_has_sref_id, "Scrim_Avg"] = (rb_df.loc[null_gms_and_has_sref_id, "Scrim_Yds"] / 
                                                rb_df.loc[null_gms_and_has_sref_id, "Plays"])
```

```python
rush_rec_per_g2 = (rb_df.loc[null_gms_and_has_sref_id, "Rush_Att":"Scrim_TD"]
                        .div(rb_df.loc[null_gms_and_has_sref_id, "G"], axis=0))
rb_df.loc[null_gms_and_has_sref_id, rush_rec_per_g_cols] = rush_rec_per_g2.values
```
